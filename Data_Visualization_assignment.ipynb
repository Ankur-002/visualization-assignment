```python
#5. Generate a synthetic dataset with correlated features. Visualize the correlation matrix of a dataset using a heatmap.

import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import numpy as np

# Set random seed for reproducibility
np.random.seed(42)

# Generate synthetic data
n_samples = 100
n_features = 5

# Create correlated features
# Start with a base feature
base_feature = np.random.normal(size=n_samples)

# Generate correlated features
data = pd.DataFrame({
    'Feature_1': base_feature,
    'Feature_2': base_feature + np.random.normal(scale=0.5, size=n_samples),
    'Feature_3': base_feature + np.random.normal(scale=0.5, size=n_samples),
    'Feature_4': base_feature + np.random.normal(scale=0.5, size=n_samples),
    'Feature_5': base_feature + np.random.normal(scale=0.5, size=n_samples)
})

# Compute the correlation matrix
matrix = data.corr()

# Create a heatmap to visualize the correlation matrix
plt.figure(figsize=(10, 8))
sns.heatmap(matrix, annot=True, cmap='coolwarm', fmt='.2f', vmin=-1, vmax=1, square=True, linewidths=0.5)

# Set plot title
plt.title('Correlation Matrix Heatmap')

# Show the plot
plt.show()
```

```python
#Using the given x and y data, generate a 3D surface plot to visualize the function z= sin((x*x +y*y))
'''x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)
z = np.sin(np.sqrt(x**2 + y**2))
data = {
    'X': x.flatten(),
    'Y': y.flatten(),
    'Z': z.flatten()
}
df = pd.DataFrame(data)'''

import plotly.graph_objects as go
import numpy as np
import pandas as pd

x = np.linspace(-5, 5, 100)
y = np.linspace(-5, 5, 100)
x, y = np.meshgrid(x, y)

z = np.sin(np.sqrt(x**2 + y**2))

fig = go.Figure(data=[go.Surface(
    z=z,
    x=x,
    y=y,
    colorscale='Viridis',
    cmin=-1,
    cmax=1
)])

fig.update_layout(
    title='3D Surface Plot of sin(sqrt(x^2 + y^2))',
    scene=dict(
        xaxis_title='X',
        yaxis_title='Y',
        zaxis_title='Z'
    )
)

fig.show()
```

```python
'''Create a Bokeh scatter plot using randomly generated x and y values. Use different sizes and colors for the
markers based on the 'sizes' and 'colors' columns.'''

output_notebook()

import numpy as np
import pandas as pd
from bokeh.plotting import figure, output_notebook, show
from bokeh.models import ColumnDataSource
from bokeh.transform import linear_cmap


np.random.seed(42)
data = {
    'x': np.random.uniform(0, 10, 100),
    'y': np.random.uniform(0, 10, 100),
    'sizes': np.random.uniform(10, 50, 100),  # Marker sizes
    'colors': np.random.uniform(0, 1, 100)    # Marker colors (for mapping)
}

df = pd.DataFrame(data)

source = ColumnDataSource(df)

p = figure(title="Scatter Plot with Varying Sizes and Colors",
           x_axis_label='x',
           y_axis_label='y',
           width=800,
           height=400)

color_mapper = linear_cmap(field_name='colors', palette='Viridis256', low=df['colors'].min(), high=df['colors'].max())

p.scatter(x='x',
          y='y',
          size='sizes',
          color=color_mapper,
          fill_alpha=0.6,
          source=source)

show(p)
```

```python
#5. Create a Bokeh heatmap using the provided dataset.
'''data_heatmap = np.random.rand(10, 10)
x = np.linspace(0, 1, 10)
y = np.linspace(0, 1, 10)
xx, yy = np.meshgrid(x, y)'''

import numpy as np
from bokeh.plotting import figure, show
from bokeh.models import ColorMapper, LinearColorMapper, BasicTicker, ColorBar

data_heatmap = np.random.rand(10, 10)


color_mapper = LinearColorMapper(palette='Viridis256', low=data_heatmap.min(), high=data_heatmap.max())

p = figure(title="Heatmap", x_axis_label='X', y_axis_label='Y')


p.image(image=data_heatmap, x=0, y=0, dw=1, dh=1,  # Adjust dw and dh for scaling if needed
        color_mapper=color_mapper)


color_bar = ColorBar(color_mapper=color_mapper, ticker=BasicTicker(), label_standoff=8)
p.add_layout(color_bar, 'right')


show(p)

```
